## SymPy’s polynomials 
- https://docs.sympy.org/latest/modules/polys/basics.html#polynomials 

$$ P_n(x) =  a_n x^n + a_{n-1} x^{n-1} + a_{n-2} x^{n-2} +\cdots +  a_1 x  + a_0$$

$$P_n(x) = 0 $$

$$a_n x^n + a_{n-1} x^{n-1} + a_{n-2} x^{n-2} +\cdots +  a_1 x  + a_0 = 0$$

$$(x-r_1) Q_{n-1}(x) = P_n(x)$$
$$(x-r_1) (x - r_2) Q_{n-2}(x) = P_n(x)$$
$$ \vdots $$

```
from sympy import Symbol
from sympy import div

x = Symbol('x')

p = x**10 - 140*x**9 + 5645*x**8 - 21516*x**7 - 1816349*x**6 - 2238724*x**5 + 116412303*x**4 + 285466780*x**3 - 2147536000*x**2 - 5346592000*x + 7096320000

p, r = div(p,  x-44)

print(p)
print(r)

p, r = div(p,  x-45)

print(p)
print(r)

p, r = div(p,  x-64)

print(p)
print(r)

p, r = div(p,  x-1)

print(p)
print(r)

```

## SymPy’s Symbolic and Numercical Complex Evaluations
- https://docs.sympy.org/latest/modules/evalf.html](https://)
- x = x1 + I*x2,y = y1 + I*y2, z = z1 + I*z2, x * y * z

<img src="https://raw.githubusercontent.com/gjhernandezp/NM/main/SymPyLab/sympylabwolfram7.jpg" />

```
from sympy import symbols, I, N, pi, expand
x1, x2, y1, y2, z1, z2 = symbols("x1 x2 y1 y2 z1 z2", real=True)  
x = x1 + I*x2
y = y1 + I*y2
z = z1 + I*z2

print(x*y*z)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))
w=1/(pi + I)
print('w=',w)
w=N(w)
print('w=',w)
```

```
from sympy import symbols, I, N, pi, expand
x3, x4, y3, y4, z3, z4 = symbols("x3 x4 y3 y4 z3 z4", real=True)  
x = x3 - I*x4
y = y3 + I*y4
z = z3 - I*z4

print(x*y*z)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))

w=N(1/(pi + I), 20)
print('w=',w)
```
